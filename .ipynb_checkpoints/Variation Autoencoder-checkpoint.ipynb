```python
import torch
import pandas as pd
import numpy as np
import torch.nn as nn
import torchvision 
import torchvision.transforms as transforms
from torchvision.utils import save_image
from torch.autograd import Variable
from torch.nn import functional as F
import matplotlib.pyplot as plt
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    plt.figure(figsize=[20,10])
    inp = inp.numpy().transpose((1, 2, 0))
    plt.imshow(inp)
```

```python
batch_size = 32
def to_img(x):
    x = 0.5 * (x + 1)
    x = x.clamp(0,1)
    x = x.view(x.size(0),3,28,28)
    return x
def vae_loss(mu, logvar,recon_x, x):
    kl_loss = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    rec_loss = F.binary_cross_entropy(recon_x, 
                                      x.view(x.size(0), -1),
                                      reduction = 'sum')
    
    return kl_loss + rec_loss
```

```python
img_trans = transforms.Compose([
    transforms.ToTensor(),
#     transforms.Normalize((0.5,),(0.5,))
])
train_data = torchvision.datasets.MNIST(root= '../data/',
                                       transform = transforms.ToTensor(),
                                       download = True)
test_data = torchvision.datasets.MNIST(root = '../data/',
                                      train = False,
                                      transform = transforms.ToTensor(),
                                      download = True)
```

```python
train_loader = torch.utils.data.DataLoader(dataset= train_data,
                                          batch_size = batch_size,
                                          shuffle = True,
                                          num_workers = 4)
test_loader = torch.utils.data.DataLoader(dataset = test_data,
                                         batch_size = batch_size,
                                         shuffle = True)
```

```python
class vae(nn.Module):
    def __init__(self):
        super(vae, self).__init__()
        self.fc1 = nn.Linear(28 * 28, 156)
        self.fc12 = nn.Linear(156, 64)
        self.fc2 = nn.Linear(156,28 * 28)
        self.fc21 = nn.Linear(64, 156)
    
    def encoder(self,x):
        x = F.relu(self.fc1(x))
        return self.fc12(x), self.fc12(x)
    
    def decoder(self,x):
        x = F.relu(self.fc21(x))
        x = F.sigmoid(self.fc2(x))
        return x
    
    def reparam(self, mu, logvar):
        std = torch.exp(0.5* logvar)
        eps = torch.randn_like(std)
        return mu + (std * eps)
    
    def forward(self, x):
        mu,logvar = self.encoder(x.view(-1, 28 * 28))
        z = self.decoder(self.reparam(mu,logvar))
        return z, mu, logvar
```

```python
learning_rate = 0.001
vae_model = vae().cuda()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(vae_model.parameters(), lr = learning_rate)

num_epochs = 4
for i in range(num_epochs):
    for data in train_loader:
        img, _ = data
        img = Variable(img)
        
        output,mu,logvar = vae_model(img.cuda())
        loss = vae_loss(mu,logvar,output, img.cuda())
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    print ('Epoch [{}/{}], Loss: {:.4f}' 
               .format(i+1, num_epochs,loss.item()))
```
